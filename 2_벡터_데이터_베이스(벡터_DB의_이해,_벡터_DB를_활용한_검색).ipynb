하이브리드 RAG
```
검색: 키워드검색, 의미기반검색(벡터DB)
RAG 튜닝 : 키워드검색,메타데이터 필터링,다중쿼리 검색 품질을 높임
BM25 : TF-IDF, 토큰화, 단어빈도 문서의 길이를 고려 점수화
결합 : 벡터검색점수 + BM25점수를 조합(가중치)
다중쿼리 : 관련쿼리를 생성해서 검색 범위 확장
  ex) 근육운동  --> 근육운동효과? 근육운동 방법?
  LangChain : MultiQueryRetiriver
성능평가 : 검색정확도, 응답시간, 요약품질(Rouge점수)
```

````python
# 하이브리드 검색 retriever
def hybrid_retrieve(query,vector_retriever,bm25_retriever,docs,k=2, vector_weight=0.7):
  # 벡터 검색
  vector_results =  vector_retriever.invoke(query)
  vector_scores =  {doc.page_content: 1/(i+1) for i, doc in enumerate(vector_results) }
  # BM25 검색
  tokenized_query = okt.morphs(query,stem=True)
  bm25_scores =  bm25.get_scores(tokenized_query)
  bm25_scores = {doc.page_content: score for doc, score in zip(docs, bm25_scores) if score > 0}
  # 점수 결합
  combined_scores = {}
  for doc in docs:
    content = doc.page_content
    v_score = vector_scores.get(content, 0)
    b_score = bm25_scores.get(content, 0)
    # vector_weight = 0.7  vector_weight * v_score--> v_score의 70퍼센트만 반영하겠다.
    # b_score 는 30% 반영
    # 가중평균 --> 합이 1되는 경우.
    combined_scores[content] = vector_weight * v_score + (1 - vector_weight) * b_score  # 가중평균 전체합이 1이되도록
  # 상위 k개 문서 선택
  sorted_docs =  sorted(combined_scores.items(), key=lambda x: x[1], reverse=True)[:k]
  return [Document(page_content=content) for content, score in sorted_docs]

from langchain.schema import BaseRetriever
from pydantic import Field
from typing import List, Any
class HybridRetriever(BaseRetriever):
    vector_retriever: Any = Field()
    bm25_retriever: Any = Field()
    docs: List[Document] = Field()
    k: int = 2
    vector_weight: float = 0.7
    def _get_relevant_documents(self, query: str) -> List[Document]:
        return hybrid_retrieve(
            query,
            self.vector_retriever,
            self.bm25_retriever,
            self.docs,
            self.k,
            self.vector_weight
        )
````

````python
hybrid_retriever = HybridRetriever(
    vector_retriever=faiss_retriever,
    bm25_retriever=bm25,
    docs=docs
)
````

````python
# LLM설정
from transformers import pipeline
from transformers import AutoModelForCausalLM, AutoTokenizer
LLM_model_id = 'naver-hyperclovax/HyperCLOVAX-SEED-Text-Instruct-0.5B'
model = AutoModelForCausalLM.from_pretrained(LLM_model_id)
tokenizer = AutoTokenizer.from_pretrained(LLM_model_id)
text_generation_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=300,
)
from langchain_huggingface import HuggingFacePipeline
llm = HuggingFacePipeline(pipeline=text_generation_pipeline)
````

````python
# 다른 llm 모델 설정해서 진행
model_id = 'Bllossom/llama-3.2-Korean-Bllossom-3B'

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto",
)
text_generation_pipeline = pipeline(
    "summarization",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=300,
)
from langchain_huggingface import HuggingFacePipeline
llm = HuggingFacePipeline(pipeline=text_generation_pipeline)
````

````python
# Rag 파이프라인 (프롬프트)
prompt_template = """
다음 문장을 50자 내외로 요약해

{context}
"""
from langchain.prompts import PromptTemplate
prompt = PromptTemplate(template=prompt_template, input_variables=["context"])
````

````python
# 하이브리드 검색( faiss)
from langchain.chains import RetrievalQA
faiss_hybrid_retriever = HybridRetriever(vector_retriever=faiss_retriever,
    bm25_retriever=bm25,
    docs=docs   )
faiss_qa =  RetrievalQA.from_chain_type(
    llm=llm,
    chain_type = "stuff",
    retriever = faiss_hybrid_retriever ,
    return_source_documents=True,
    chain_type_kwargs = {"prompt": prompt}
    )

chroma_hybrid_retriever = HybridRetriever(vector_retriever=chroma_retriever,
    bm25_retriever=bm25,
    docs=docs   )
chroma_qa =  RetrievalQA.from_chain_type(
    llm=llm,
    chain_type = "stuff",
    retriever = chroma_hybrid_retriever ,
    return_source_documents=True,
    chain_type_kwargs = {"prompt": prompt}
    )
````

````python
# 요약
query = "요약해줘"
faiss_summary = faiss_qa.invoke(query)
````

````python
# 모델 검증

from transformers import AutoModelForCausalLM, AutoTokenizer
model = AutoModelForCausalLM.from_pretrained("naver-hyperclovax/HyperCLOVAX-SEED-Text-Instruct-0.5B")
tokenizer = AutoTokenizer.from_pretrained("naver-hyperclovax/HyperCLOVAX-SEED-Text-Instruct-0.5B")

mydocs = """
'서울아산병원 연구팀이 심혈관질환 환자가 중강도 이상의 운동을 꾸준히 할 경우 질환 재발 위험이 13%까지 낮아진다는 연구 결과를 발표했다.',
"강남구가 '2025 건강 뜀' 사업을 통해 청소년 체력 증진, 중년 여성 비만 예방 등 전 세대를 위한 맞춤형 운동 프로그램을 운영하고 있다."
"""
chat = [
  {"role": "tool_list", "content": ""},
  {"role": "system", "content": "- AI 언어모델의 이름은 \"CLOVA X\" 이며 네이버에서 만들었다.\n- 오늘은 2025년 05월 13일(화)이다."},
  {"role": "documents", "content": mydocs},
  {"role": "user", "content": "문장을 보고 관련 자료를 만들어줘"},
]

inputs = tokenizer.apply_chat_template(chat, add_generation_prompt=True, return_dict=True, return_tensors="pt")
output_ids = model.generate(**inputs, max_length=1024, stop_strings=["<|endofturn|>", "<|stop|>"], tokenizer=tokenizer)
````
