```python
number_1 = input("적절한 숫자를 입력하세요 : ")

# 정수일 경우
if number_1.isdigit():
    number_1 = int(number_1)
# 실수일 경우
elif '.' in number_1:
    try:
        number_1 = float(number_1)
    except ValueError:
        print("유효한 숫자가 아닙니다.")
else:
    print("유효한 숫자가 아닙니다.")

print(f"입력한 숫자: {number_1}")
```
